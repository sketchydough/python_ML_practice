```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Boston.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
y=['MEDV']
X=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']
```

```python
#model(linear regression) using statsmodels.api library
import statsmodels.api as sm
```

```python
#OLS-> ordinary least squares. .fit()-> estimates the coeffs of regression. .summary()-> prints statistical report for regression
print(sm.OLS(df[y], df[X]).fit().summary())
```
