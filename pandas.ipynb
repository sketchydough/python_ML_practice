```python
#lets take two diff datasets to perform few data analyzing operations on them

import pandas as pd

df1= pd.read_csv('time_series_covid_19_recovered.csv')
df2= pd.read_csv('time_series_covid_19_deaths.csv')
```

```python
#the total number of patients

df1['Country/Region'].value_counts().sum()
```

```python
#total number of patients who are unique
df1['Country/Region'].nunique()
```

```python
#merge the datasets

merged_df= pd.merge(df1,df2, on='Country/Region',how='inner')
```

```python
merged_df.head()
```

```python
merged_df.info()
```

```python
#check for missing values

merged_df.isnull().sum()
```

```python
# we are not fixing the missign values because we dont have anything to manipulate at country nor the lat or long
```

```python
#problem: let say same day multiple patients from the same country. so how many in total?

summary_day= merged_df.groupby('Country/Region')['5/27/21_y'].sum() # here group by country is where it checks no of multiple entries/patients and the date is where the sum is being performed

summary_day
```

```python
# problem: let say again multiple patients from the same country but on diff days for a whole week data

#using aggregate->sum/avg,etc we can perform dictionary format

summary_week= merged_df.groupby('Country/Region').agg({'5/25/21_y':'sum', '5/26/21_y':'sum','5/27/21_y':'sum','5/28/21_y':'sum','5/29/21_y':'sum'})

summary_week
```

```python
#problem: Filtering patients->if the no of patients is above 2000

filtered_summary= merged_df[merged_df['5/27/21_y']>2000]
filtered_summary
```

```python
#Plotting

#Problem: getting the plot of only a particular day in week summary of data

summary_week['5/25/21_y'].plot( kind='bar', figsize=(30,10), title=' Total no of patients by country/region', color='lightgreen')
```
