```python
# import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.array(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
sales= np.array([200,230,45,567,577,13,245,456,876,245,267,654])
```

```python
#basic line plot

plt.figure(figsize=(6,4))
plt.plot(months,sales,marker='o',linestyle='-',color='b') #note:plt.plot
plt.title('monthly sales data')
plt.xlabel('month')
plt.ylabel('sales')
plt.grid(True)
plt.show()
```

```python
#histogram

#create a dataset for histogram
np.random.seed(0)
data=np.random.normal(loc=0,scale=1,size=1000) #generate random data w normal distribution

plt.figure(figsize=(6,4))
plt.hist(data,bins=30,color='purple',edgecolor='black') #note: plt.hist
plt.title('random')
plt.xlabel('mvalue')
plt.ylabel('freq')
plt.grid(True)
plt.show()
```

```python
#scatter plot

np.random.seed(1)
x=np.random.rand(50)*100
y=np.random.rand(50)*100

plt.figure(figsize=(6,4))
plt.scatter(x,y,color='purple',alpha=0.7) #note: plt.scatter
plt.title('random')
plt.xlabel('mvalue')
plt.ylabel('freq')
plt.grid(True)
plt.show()
```

```python
#bar plot

categories=np.array(['a','b','c','d','e'])
values=np.array([10,20,30,40,50])

plt.figure(figsize=(6,4))
plt.bar(categories,values,color=['blue','orange','red','pink','purple']) #note: plt.bar
plt.title('random')
plt.xlabel('mvalue')
plt.ylabel('freq')
plt.grid(True)
plt.show()
```

```python
#Seaborn
import pandas as pd
import seaborn as sns

import warnings

warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('covid_19_data.csv')
```

```python
df.head()
```

```python
sns.set(style='whitegrid') #use white grid style for better visualization
```

```python
#histogram

plt.figure(figsize=(6,4))
sns.histplot(df['SNo'],bins=20,kde=True,color='purple')
plt.title('random')
plt.xlabel('mvalue')
plt.ylabel('freq')
plt.grid(True)
plt.show()
```

```python
#scatter plot with hue

plt.figure(figsize=(6,4))
sns.scatterplot(x='Confirmed',y='Deaths',hue='Country/Region',data=df,palette='deep',alpha=0.7)
plt.title('random')
plt.xlabel('mvalue')
plt.ylabel('freq')
plt.grid(True)
plt.show()
```

```python
#pair plot

plt.figure(figsize=(5,5))
sns.pairplot(df[['Recovered','Deaths','Confirmed']],diag_kind='kde',hue='Confirmed')
plt.show()
```
